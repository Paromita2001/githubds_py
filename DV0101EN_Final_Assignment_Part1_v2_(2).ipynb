### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# Create dataframes for recession and non-recession period
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# Figure
fig = plt.figure(figsize=(12, 6))

# Create different axes for subplotting
ax0 = fig.add_subplot(1, 2, 1)  # subplot 1 (recession)
ax1 = fig.add_subplot(1, 2, 2)  # subplot 2 (non-recession)

# Plot GDP variation during recession
sns.lineplot(x='Year', y='GDP', data=rec_data, label='Recession', ax=ax0)
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during Recession Period')

# Plot GDP variation during non-recession
sns.lineplot(x='Year', y='GDP', data=non_rec_data, label='Non-Recession', ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during Non-Recession Period')

plt.tight_layout()
plt.show()
````

 ### TASK 1.7: Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.
 <br>How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods?


<details><summary>Click here for a hint</summary>

<p>
     You can create two dataframe for recession and nonreccession period.
    <br> Calculate the sum of Advertising_Expenditure for both dataframes
    <br> Pass these total values to plt.pie(). May include labels as ['Recession', 'Non-Recession']
        <br>Feel Free to customie the pie further
    <br>title this plot as  - Advertising Expenditure during Recession and Non-Recession Periods</p>
</details>


````python
# Filter the data
Rdata = df[df['Recession'] == 1]
NRdata = df[df['Recession'] == 0]

# Calculate total advertising expenditure for both periods
RAtotal = Rdata['Advertising_Expenditure'].sum()
NRAtotal = NRdata['Advertising_Expenditure'].sum()

# Create a pie chart
plt.figure(figsize=(8, 6))
labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Advertising Expenditure during Recession and Non-Recession Periods')
plt.show()
````

````python
# TASK 1.1  Line_plot_1.png
import pandas as pd
import matplotlib.pyplot as plt

# Group by year to get average sales
yearly_sales = df.groupby('Year', as_index=False)['Automobile_Sales'].mean()

# Create a line plot
plt.figure(figsize=(10,6))
plt.plot(yearly_sales['Year'], yearly_sales['Automobile_Sales'], marker='o', color='blue')

plt.title('Yearly Automobile Sales Trend')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.grid(True, linestyle='--', alpha=0.5)

# Save image
plt.savefig('Line_plot_1.png')
plt.show()
````

````python
# TASK 1.2  Line_plot_2.png

# Filter data for recession period
recession_data = df[df['Recession'] == 1]

# Group by year and vehicle type, get mean sales
recession_group = recession_data.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Plot different lines for each vehicle type
plt.figure(figsize=(10,6))
for vehicle in recession_group['Vehicle_Type'].unique():
    subset = recession_group[recession_group['Vehicle_Type'] == vehicle]
    plt.plot(subset['Year'], subset['Automobile_Sales'], marker='o', label=vehicle)

plt.title('Automobile Sales Trend During Recession by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Save image
plt.savefig('Line_plot_2.png')
plt.show()
````

````python
# TASK 1.3  Bar_Chart.png
import seaborn as sns

# Group and average sales by Recession and Vehicle_Type
avg_sales = df.groupby(['Recession', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=avg_sales)

plt.title('Average Automobile Sales per Vehicle Type (Recession vs Non-Recession)')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Recession (1=Yes, 0=No)')
plt.grid(True, linestyle='--', alpha=0.5)

# Save image
plt.savefig('Bar_Chart.png')
plt.show()
````

````python
from dash import Dash, html, dcc, Input, Output
import pandas as pd
import plotly.express as px

# Load your dataset
df = pd.read_csv(URL)

# Create app
app = Dash(__name__)

# App layout
app.layout = html.Div([
    html.H1("Automobile Data Dashboard", style={'textAlign': 'center'}),

    html.Label("Select Report Type:", style={'fontSize': 18}),
    dcc.Dropdown(
        id='report-type-dropdown',
        options=[
            {'label': 'Recession Report', 'value': 'recession'},
            {'label': 'Yearly Report', 'value': 'yearly'},
        ],
        value='recession'
    ),

    html.Div(id='output-container', style={'marginTop': 30})
])

# ----------------------------------------------------------
# Define Callback Function
# ----------------------------------------------------------
@app.callback(
    Output('output-container', 'children'),
    Input('report-type-dropdown', 'value')
)
def update_output(selected_report):
    if selected_report == 'recession':
        filtered_df = df[df['Recession'] == 1]
        fig = px.line(
            filtered_df,
            x='Year',
            y='Automobile_Sales',
            color='Vehicle_Type',
            title='Recession Report: Automobile Sales Trend'
        )
    else:
        filtered_df = df.groupby('Year')['Automobile_Sales'].mean().reset_index()
        fig = px.bar(
            filtered_df,
            x='Year',
            y='Automobile_Sales',
            title='Yearly Report: Average Automobile Sales'
        )

    return dcc.Graph(figure=fig)

# ----------------------------------------------------------
# Run the Dash app
# ----------------------------------------------------------
if __name__ == '__main__':
    app.run(debug=True)
````
