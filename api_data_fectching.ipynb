```python
import pandas as pd
import numpy as np
import requests
```

```python
url1='https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1'
response=requests.get(url1)
df=pd.DataFrame(response.json()['results'])[['title','overview']]
```

```python
url2='https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US'
response=requests.get(url2)
data=pd.DataFrame(response.json()['genres'])
```

```python
df=pd.DataFrame()
for i in range(1,501):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    if 'results' in response.json():
        df1=pd.DataFrame(response.json()['results'])[['title','overview','genre_ids']]
        df=pd.concat([df,df1],ignore_index=True)
```

```python
df
```

```python
gen_df=dict(zip(data['id'],data['name']))
```

```python
def convert_id_name(genre_ids):
  return[gen_df.get(i,'unknown') for i in genre_ids]
```

```python
df['genre_names']=df['genre_ids'].apply(convert_id_name)
```

```python
df
```

```python
df['genre']=df['genre_names'].apply(lambda x: ','.join(x))
```

```python
df.drop(['genre_ids','genre_names'],inplace=True , axis=1)
```
