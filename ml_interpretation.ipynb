## **1.(c) Evaluate the covariance and correlation matrices to analyze the relationships between the variables.**

```python
# Covariance
print("Covariance Matrix:")
df.cov()
```

```python
# Correlation
print("Correlation Matrix:")
df.corr()
```

**Insights**

Murder ~ Assault: These have a high correlation of (~0.80), with a covariance of 291.06. States with high assault rates also tend to report more murders. This suggests a strong linear dependency between these two crimes.

Murder ~ UrbanPop: The correlation is very low (~0.07) and the covariance is 4.38, showing there is almost no linear relationship between the population of urban area and murder rates. This means urbanization does not significantly influence murder arrests.

Murder ~ Rape: Shows a moderate correlation (~0.56) and a covariance of 22.99, indicating that states with higher murder rates may also report more rape cases. This might mean that similar social or safety problems are causing both types of crimes

Assault ~ UrbanPop: With a correlation of ~0.26, with a covariance of 312.27, There isn't a clear link between assault arrests and the size of the city. Meaning, more people living in urban areas doesn't always lead to more assaults.

Assault ~ Rape: A strong positive correlation (~0.66) with a covariance of 519.26. If a state has a lot of assault cases, it likely to has more rape cases too. This could be because of similar problems or patterns in how crimes are reported.

UrbanPop ~ Rape: The correlation is moderate (~0.41) and the covariance is 55.77. Urban areas may have more reported rape cases than rural areas, but this relationship is not very strong.

```python
sns.pairplot(df)
plt.suptitle("Scatterplot Matrix of USArrests", y=1.02)
plt.show()
```
